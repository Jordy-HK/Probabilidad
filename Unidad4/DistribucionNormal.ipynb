## Ejercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desv_std = 4
num_datos = 500
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50]
```

```python
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)
pdf = 1/(desv_std*np.sqrt(2*np.pi))*np.exp(-(x-media)**2/(2*desv_std**2))
print("Imprimir los primeros 50 Valores de x", x[0:50], ", y los primeros 50 pdf", pdf[0:50])
```

```python
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal: (media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Pesos en kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2,)
plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

## Ejercicio 2

```python
plt.figure(figsize=(6, 4))
media = 70
desv_std = 4
total_estudiantes = 500
peso_limite = 75
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg
print('Probabilidad de que un estudiante pese menos de 75 kg:', prob_menos_75kg)
print('Numero de estudiantes que pesan menos de 75 kg:', estudiantes_menos_75kg)
x = np.linspace(media-4*desv_std, media+4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite),color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
media = 70
desv_std = 4
total_estudiantes = 500
peso_limite = 62
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg
print('Probabilidad de que un estudiante pese menos de 62 kg:', prob_menos_62kg)
print('Numero de estudiantes que pesan menos de 62 kg:', estudiantes_menos_62kg)
```

```python
x = np.linspace(media-4*desv_std, media+4*desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite),color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
media = 70
desv_std = 4
total_estudiantes = 500
peso_limite = 79
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)
prob_mas_79kg = 1-prob_menos_79kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg
print('Probabilidad de que un estudiante pese mas de 79 kg:', prob_mas_79kg)
print('Numero de estudiantes que pesan menos de 79 kg:', estudiantes_mas_79kg)
```

```python
x = np.linspace(media-4*desv_std, media+4*desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite),color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 79 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```
