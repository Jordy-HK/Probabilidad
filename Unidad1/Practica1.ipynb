```python
from statistics import *
grupoa = [20, 87, 98, 99, 100, 100, 50, 68, 70, 80, 85, 96,]
print(grupoa)
type(grupoa)
promedio = sum(grupoa)/len(grupoa)
print(promedio)

promedio2 = mean(grupoa)
print(promedio2)

mediana = median(grupoa)
print(mediana)
moda = mode(grupoa)
print(moda)
moda2 = multimode(grupoa)
print(moda2)
```

```python
tiempos100mts = [15.10, 12.30, 14.11, 11.02, 20.14, 10.01, 9.99, 9.89, 16.25, 11.98]
print(tiempos100mts)
print("El tiempo promedio de los corredores es: ", mean(tiempos100mts))
print("El tiempo mediano de los corredores es: ", median(tiempos100mts))
print("El tiempo de la moda de los corredores es: ", mode(tiempos100mts))
```

```python
import random
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)

fabrica2 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
import matplotlib.pyplot as plt
plt.hist(fabrica2)
plt.title("Produccion diarria de la fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
print("Valores de la fabrica 1")
print("La media es: ", media1)
print("La median es: ", mediana1)
print("La moda es: ", moda1)

media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("Valores de la fabrica 2")
print("La media es: ", media2)
print("La median es: ", mediana2)
print("La moda es: ", moda2)
```
