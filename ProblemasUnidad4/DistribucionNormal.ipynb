### **Problema de Distribucion Normal**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
plt.figure(figsize=(6, 4))
media = 70
desv_std = 4
total_estudiantes = 500
peso_minimo = 68
peso_limite = 72
prob_menos_72kg = norm.cdf(peso_limite, media, desv_std) - norm.cdf(peso_minimo, media, desv_std)
prob_mas_68kg = 1 - prob_menos_72kg


estudiantes_menos_72kg = total_estudiantes * prob_menos_72kg
estudiantes_mas_68kg = total_estudiantes - estudiantes_menos_72kg

print('Probabilidad de que un estudiante pese menos de 72 kg:', prob_menos_72kg)
print('Numero de estudiantes que pesan menos de 72 kg:', estudiantes_menos_72kg)
print('Probabilidad de que un estudiante pese mas de 68 kg:', prob_mas_68kg)
print('Numero de estudiantes que pesan mas de 68 kg:', total_estudiantes - estudiantes_mas_68kg)
x = np.linspace(media-4*desv_std, media+4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_minimo) & (x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68 kg y menos de 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```
