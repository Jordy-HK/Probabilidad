```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
```

```python
autos = pd.read_csv("autos_corregidos.csv")
autos.head()
```

```python
autos.info()
```

```python

autos.price.max()
```

```python
autos.price.min()
```

```python
media=mean(autos.price)
mediana=median(autos.price)
moda=mode(autos.price)
desviacion_estandar_autos = stdev(autos.price)
print("Precio de los Autos según: \n", "media =", media,
      "\n mediana = ", mediana, "\n moda = ", moda)
print("La desviacion estandar es: ", desviacion_estandar_autos)
```

```python
import math
import matplotlib.pyplot as plt
df = pd.DataFrame({'autos.price': autos.price})
df
n = len('autos.price')
k = int(1 + math.log2(n))
print("Valor de n= ", n, "\nNumero de intervalos=", k)
plt.hist(df['autos.price'], bins=k, alpha = 0.5)
plt.title('Histograma de Precio de Autos basado en la regla de Sturges')
plt.xlabel('Precio')
plt.ylabel('Frecuencia de compra')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='body-style', y='price', data=autos)

plt.title("Relación Precio y Tipo de Vehículo", fontsize=14)
plt.xlabel("Tipo de Vehículo", fontsize=12)
plt.ylabel("Precio del Automóvil", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='engine-location', y='price', data=autos)

plt.title("Relación Precio del Auto y Ubicación del Motor", fontsize=14)
plt.xlabel("Ubicación del Motor", fontsize=12)
plt.ylabel("Precio del Automóvil", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='drive-wheels', y='price', data=autos)

plt.title("Relación Precio del Automóvil y Tipo de Tracción", fontsize=14)
plt.xlabel("Tipo de Tracción", fontsize=12)
plt.ylabel("Precio del Automóvil", fontsize=12)

plt.show()
```
