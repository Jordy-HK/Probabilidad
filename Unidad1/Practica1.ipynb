```python
from statistics import *
grupoa = [20, 87, 98, 99, 100, 100, 50, 68, 70, 80, 85, 96,]
print(grupoa)
type(grupoa)
promedio = sum(grupoa)/len(grupoa)
print(promedio)

promedio2 = mean(grupoa)
print(promedio2)

mediana = median(grupoa)
print(mediana)
moda = mode(grupoa)
print(moda)
moda2 = multimode(grupoa)
print(moda2)
```

```python
tiempos100mts = [15.10, 12.30, 14.11, 11.02, 20.14, 10.01, 9.99, 9.89, 16.25, 11.98]
print(tiempos100mts)
print("El tiempo promedio de los corredores es: ", mean(tiempos100mts))
print("El tiempo mediano de los corredores es: ", median(tiempos100mts))
print("El tiempo de la moda de los corredores es: ", mode(tiempos100mts))
```
