```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
```

```python
autos = pd.read_csv("autos_corregidos.csv")
autos.head()
```

```python
autos.info()
```

```python

autos.price.max()
```

```python
autos.price.min()
```

```python
media=mean(autos.price)
mediana=median(autos.price)
moda=mode(autos.price)
desviacion_estandar_autos = stdev(autos.price)
print("Precio de los Autos según: \n", "media =", media,
      "\n mediana= ", mediana, "\n moda= ", moda)
print("La desviacion estandar es: ", desviacion_estandar_autos)
```

```python
media = autos.price.mean()
plt.pyplot.hist(autos.price, bins = 20)
plt.pyplot.title("Histograma de Precio de Autos")
plt.pyplot.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Frecuencia")
```
